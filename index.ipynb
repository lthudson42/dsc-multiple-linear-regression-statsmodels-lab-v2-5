## The Ames Housing Dataset

The [Ames Housing dataset](http://jse.amstat.org/v19n3/decock.pdf) is a newer (2011) replacement for the classic Boston Housing dataset. Each record represents a residential property sale in Ames, Iowa. It contains many different potential predictors and the target variable is `SalePrice`.

````python
import pandas as pd
ames = pd.read_csv("ames.csv", index_col=0)
ames
````

````python
ames.describe()
````

We will focus specifically on a subset of the overall dataset. These features are:

```
LotArea: Lot size in square feet

1stFlrSF: First Floor square feet

GrLivArea: Above grade (ground) living area square feet
```

````python
ames_subset = ames[['LotArea', '1stFlrSF', 'GrLivArea', 'SalePrice']].copy()
ames_subset
````

## Step 1: Visualize Relationships Between Features and Target

For each feature in the subset, create a scatter plot that shows the feature on the x-axis and `SalePrice` on the y-axis.

````python
# Your code here - import relevant library, create scatter plots
import matplotlib.pyplot as plt
import seaborn as sns

fig, axes = plt.subplots(1, 3, figsize=(15, 5))
axes[0].set_ylabel("Sale Price")

for i, col in enumerate(ames_subset.drop("SalePrice", axis = 1).columns):
    ax = axes[i]
    ax.scatter(ames_subset[col], ames_subset["SalePrice"])
    ax.set_xlabel(col)
    
plt.show();
````

````python
# Your written answer here - do these seem like good candidates for linear regression?
"""
Yes, these features seem like good candidates for linear regression because they show a linear relationship with the target varaiable, SalePrice.
"""
````

## Step 2: Build a Simple Linear Regression Model

Set the dependent variable (`y`) to be the `SalePrice`, then choose one of the features shown in the subset above to be the baseline independent variable (`X`).

Build a linear regression using StatsModels, describe the overall model performance, and interpret its coefficients.

````python
# Your code here - define y and baseline X
y = ames_subset["SalePrice"]
X = ames_subset["GrLivArea"]
````

````python
# Your code here - import StatsModels, fit baseline model, display results
import statsmodels.api as sm

baseline_model = sm.OLS(y, sm.add_constant(X))
baseline_results = baseline_model.fit()

print(baseline_results.summary())
````

$$ \hat{y} = \hat{\beta_0} + \hat{\beta_1} x $$

````python
# Your written answer here - interpret model results
'''
Given the above formula for a simple linear regression and the results of the baseline model, we can interpret the results as follows:

- The coefficient for GrLivArea is 107.1304. This means that for each additional square foot of above ground living area, we expect the SalePrice to increase by $107.13.
- The const value is 18570. This means that if the above ground living area is 0, we expect the SalePrice to be $18,570.
- GrLivArea is statistically significant at the 0.05 level, as the t-statistic p-value is less than 0.05.
- The R-squared value is 0.502, which means that 50.2% of the variance in SalePrice can be explained by the model.
'''
````

## Step 3: Build a Multiple Linear Regression Model

For this model, use **all of** the features in `ames_subset`.

````python
# Your code here - define X
X_multi = ames_subset[['LotArea', '1stFlrSF', 'GrLivArea']]
````

````python
# Your code here - fit model and display results
second_model = sm.OLS(y, sm.add_constant(X_multi))
second_results = second_model.fit()

print(second_results.summary())
````

````python
# Your written answer here - interpret model results. Does this model seem better than the previous one?
'''
- The model explains 56.5% of the variance in SalePrice.
- The const value is -14310. This means that if all the features are 0, we expect the SalePrice to be -$14,310. This is not a meaningful interpretation because we cannot have a negative Sale Price.
- All feature variables are statistically significant at the 0.05 level, except for LotArea, which has a t-statistic p-value of 0.051.
- For every 1 square foot increase in GrLivArea, we expect an $80.6061 increase in Sale Price.
- For every 1 square foot increase in 1stFlrSF, we expect a $60.2866 increase in Sale Price.
- For every 1 square foot increase in LotArea, we expect a $0.2841 increase in Sale Price.
'''
````

## Step 4: Create Partial Regression Plots for Features

Using your model from Step 3, visualize each of the features using partial regression plots.

````python
# Your code here - create partial regression plots for each predictor
fig = plt.figure(figsize=(10,5))

sm.graphics.plot_partregress_grid(second_results, exog_idx=['LotArea', '1stFlrSF', 'GrLivArea'], fig=fig)
plt.tight_layout()
plt.show();
````

````python
# Your written answer here - explain what you see, and how this relates
# to what you saw in Step 1. What do you notice?

'''
These partial regression plots display the parts of the feature variables (i.e., LotArea, 1stFlrSF, and GrLivArea) and the parts of SalePrice that are not explained by the model.
Since the plots of GrLivArea and 1stFlrSF show a linear relationship with a non-zero slope, this suggests that it is beneficial to add these features to the model,
as opposed to having a model without them.
'''
````
